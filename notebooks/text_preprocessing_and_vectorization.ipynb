```python
def skills_processing(x):
    skills=[]
    skill_arr=x.split(",")
    for skill in skill_arr:
        skills.append("".join(skill.split()))
    return ", ".join(skills).strip()
```

```python
df['Skills']=df['Skills'].apply(lambda x:skills_processing(x))
df.head()
```

```python
def job_title_processing(x):
    return "".join(x.split(" ")).strip()
```

```python
df['Job_title']=df['Job_title'].apply(lambda x:job_title_processing(x))
df.head()
```

```python
def experience_processing(x):
    range_of_years=x.split()[0]
    avg_years_of_exp_req=(float(range_of_years.split("-")[0])+float(range_of_years.split("-")[1]))/2
    if avg_years_of_exp_req > 0 and avg_years_of_exp_req <= 2:
        return "Entry"
    elif avg_years_of_exp_req > 2 and avg_years_of_exp_req <= 6:
        return "Associate"
    elif avg_years_of_exp_req > 6 and avg_years_of_exp_req <= 10:
        return "MidSenior"
    elif avg_years_of_exp_req > 10 and avg_years_of_exp_req <= 16:
        return "Senior"
    elif avg_years_of_exp_req > 16:
        return "Executive"
```

```python
df['Experience']=df['Experience'].apply(lambda x: experience_processing(x))
df.head()
```

## Text Preprocessing

```python
!python -m spacy download en_core_web_sm
```

```python
import string
import spacy
from nltk.stem.porter import PorterStemmer
exclude=string.punctuation
tokenizer=spacy.load('en_core_web_sm')


def text_preprocessing(text):
    # Lowercasing
    text=str(text)
    text=text.lower() 

    # Removing punctuations
    
    text=text.translate(str.maketrans('','',exclude))

    # Tokenization
    text_new=[]
    for i in text.split(): 
        text_new.append(i.strip())  
    token_list=list(tokenizer(" ".join(text_new)))  

    # Stemming
    for i in range(0,len(token_list)):
        token_list[i]=PorterStemmer().stem(str(token_list[i]))
    
    return " ".join(token_list)



```

```python
df['Tags']=df['Tags'].apply(lambda x: text_preprocessing(x))
df.head()
```

```python
job_title=input("Enter your desired job profile:")
skills=input("Enter at least 4 skills")
experience=input("Enter your level of experience")
```

```python
def recommend(job_title,skills,experience):
    job_title="".join(job_title.split())
    df2=pd.DataFrame({'Location':['Not Specified'],'Tags':[job_title+" "+skills+" "+experience]})
    df2['Tags']=df2['Tags'].apply(lambda x: text_preprocessing(x))
    df3=pd.concat([df2,df],axis=0)
    cv2=CountVectorizer()
    vectors2=cv2.fit_transform(df3['Tags']).toarray()
    similarity2=cosine_similarity(vectors2)
    top_5_recommended=sorted(list(enumerate(similarity2[0])),reverse=True,key=lambda x:x[1])[1:6]
    list_of_index_values=[]
    for item in top_5_recommended:
        list_of_index_values.append(item[0])
    return ref_df.iloc[list_of_index_values]
    
    
    
    
    
    
    
    
    
```

```python
recommend(job_title,skills,experience)
```
